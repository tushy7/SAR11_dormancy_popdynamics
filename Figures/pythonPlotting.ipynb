```python
#This is before dilutions were added

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("heatMapData1.0.csv")

# rows = phi, columns = dt, values = q
heatmap_data = data.pivot(index="phi", columns="dt", values="q")

plt.figure(figsize=(6, 4))
ax = sns.heatmap(heatmap_data, cmap="viridis", cbar_kws={'label': 'Quiescence rate'})
plt.xlabel("# days between dilution")
plt.ylabel("Virus adsorption rate")

ax.set_xticklabels([f"{tick:.2f}" for tick in heatmap_data.columns], rotation=45)
ax.set_yticklabels([f"{tick:.1e}" for tick in heatmap_data.index], rotation=0)

plt.title("Quiescence rate")
plt.tight_layout()
plt.savefig("Heat_Map.tiff")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("growth_with_dormancy.csv")
total_cells = data["Active"] + data["Quiescent"] + data["Infected_Quiescent"]

plt.plot(data["time"], total_cells, label = "Total Cells", color = "#377eb8")
plt.plot(data["time"], data["Active"], label = "Active Cell", color = "#ff7f00", linestyle = "--")
plt.plot(data["time"], data["Viruses"], label = "Viruses", color = "#4daf4a", linestyle = "--")

plt.ylim(10e-10, 10e5)
plt.yscale("log")

plt.xlabel("Time (Days)", fontsize = 11)
plt.ylabel("Cell Density (pmol cellular C/mL)", fontsize = 11)
plt.legend(loc = "lower right")

plt.savefig("SAR11_Dynamics_with_Dormancy.tiff")
plt.show()
```
