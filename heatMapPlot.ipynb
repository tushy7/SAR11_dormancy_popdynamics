```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("heatMapData.csv")

# rows = phi, columns = dt, values = q
heatmap_data = data.pivot(index="phi", columns="dt", values="q")

plt.figure(figsize=(6, 4))
ax = sns.heatmap(heatmap_data, cmap="viridis", cbar_kws={'label': 'Quiescence rate'})
plt.xlabel("# days between dilution")
plt.ylabel("Virus adsorption rate")

ax.set_xticklabels([f"{tick:.2f}" for tick in heatmap_data.columns], rotation=45)
ax.set_yticklabels([f"{tick:.1e}" for tick in heatmap_data.index], rotation=0)

plt.title("Quiescence rate")
plt.tight_layout()
plt.show()
```
